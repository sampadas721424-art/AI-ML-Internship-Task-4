```python
import pandas as pd

df = pd.read_csv('adult.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['income'] = le.fit_transform(df['income'])
```

Label Encoding is used where categories have an inherent order.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['age', 'hours-per-week', 'capital-gain', 'capital-loss']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

Label Encoding is used where categories have an inherent order.

```python
df.columns
```

```python
df = pd.get_dummies(
    df,
    columns=['gender', 'workclass', 'marital-status'],
    drop_first=True
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['age', 'hours-per-week', 'capital-gain', 'capital-loss']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

Scaling ensures all numerical features are on the same scale.

Before scaling:

Features had different ranges

After scaling:

Mean ≈ 0

Standard deviation ≈ 1

Scaling improves performance of distance-based algorithms like KNN, SVM, and Logistic Regression.

```python
df.to_csv('adult_processed.csv', index=False)
```
